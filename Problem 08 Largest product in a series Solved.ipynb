```python
def adjacent_digits(length, number):
    start_number = 0
    end_number = int(length)
    product = 0
    while end_number <= len(number):
        current_slice = number[start_number:end_number]
        current_product = functools.reduce(lambda x, y: x*y, [int(i) for i in current_slice])
        if current_product > product:
            product = current_product
        start_number += 1
        end_number += 1
    return product
```
